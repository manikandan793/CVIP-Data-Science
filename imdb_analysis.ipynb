```python
data['Gross'] = data['Gross'].str.replace(',','')
```

```python
data['Gross'] = data['Gross'].replace(np.nan, 0)
```

```python
data['Gross'] = data['Gross'].astype(int)
```

```python
data.corr()
```

```python
numerical_cols = [col for col in data.columns if data[col].dtype != 'object']
categorical_cols = [col for col in data.columns if data[col].dtype == 'object']
```

```python
data[numerical_cols].hist(figsize = (15, 6), color = 'yellow', edgecolor = 'black', layout = (2, 2))
```

```python
top_rated = data.sort_values(['No_of_Votes'], ascending = False)
```

```python
# Top voted movies
fig,axs=plt.subplots(figsize=(15,5))
g=sns.barplot(x=top_rated['Series_Title'][:7],y=top_rated['No_of_Votes'][:7], palette = 'hls')
g.set_title("Top Voted Movies", weight = "bold")
plt.show()
```

```python
# Top voted movie ratings
fig,axs=plt.subplots(figsize=(15,5))
g=sns.barplot(x=top_rated['Series_Title'][:7],y=top_rated['IMDB_Rating'][:7], palette = 'husl')
g.set_title("IMDB Rating of top voted movies", weight = "bold")
plt.show()
```

```python
# Top voted movies metascores
fig,axs=plt.subplots(figsize=(15,5))
g=sns.barplot(x=top_rated['Series_Title'][:7],y=top_rated['Meta_score'][:7], palette = 'husl')
g.set_title("Metascore of top rated movies", weight = "bold")
plt.show()
```

```python
# Gross of top rated movies
fig,axs=plt.subplots(figsize=(15,5))
g=sns.barplot(x=top_rated['Series_Title'][:7],y=top_rated['Gross'][:7], palette = 'husl')
g.set_title("Gross by top rated movies", weight = "bold")
plt.show()
```

```python
# Top movies by Gross
highest_Gross = data.sort_values(['Gross'], ascending = False)
fig,axs=plt.subplots(figsize=(23,5))
g=sns.barplot(x=highest_Gross['Series_Title'][:7],y=highest_Gross['Gross'][:7], palette = 'hls')
g.set_title("Movies with highest Gross", weight = "bold")
plt.show()
```

```python
fig,axs=plt.subplots(figsize=(20,5))
g=sns.barplot(x=data['Released_Year'].value_counts()[:10].index,y=data['Released_Year'].value_counts()[:10])
g.set_title("Maximum Series released in-", weight = "bold")
g.set_xlabel("Years")
plt.show()
```
